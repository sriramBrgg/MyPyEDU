```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/Premalatha-success/eduonix_batch1/main/Automobile.csv"
mpg_df = pd.read_csv(url)
```

```python
print(mpg_df.head())
print(mpg_df.shape)
print(mpg_df.describe())
print(mpg_df.dtypes)
```

```python
print(mpg_df.isnull().sum())
mpg_df = mpg_df.dropna() 
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=mpg_df, x='horsepower', y='highway_mpg')
plt.title('Relationship between Horsepower and Highway_mpg')
plt.xlabel('Horsepower')
plt.ylabel('highway_mpg')
plt.show()
```

```python
plt.figure(figsize=(12, 10))

# Subplot 1: Histogram of Acceleration
plt.subplot(2, 2, 1)
sns.histplot(data=mpg_df, x='horsepower', kde=True)
plt.title('Histogram of Horsepower')
plt.xlabel('Horsepower')

# Subplot 2: Histogram of Displacement
plt.subplot(2, 2, 2)
sns.histplot(data=mpg_df, x='peak_rpm', kde=True)
plt.title('Histogram of peak_rpm')
plt.xlabel('peak_rpm')

# Subplot 3: Histogram of Weight
plt.subplot(2, 2, 3)
sns.histplot(data=mpg_df, x='curb_weight', kde=True)
plt.title('Histogram of curb_weight')
plt.xlabel('curb_weight')

# Subplot 4: Histogram of Kilometer_per_litre
plt.subplot(2, 2, 4)
sns.histplot(data=mpg_df, x='city_mpg', kde=True)
plt.title('Histogram of city_mpg')
plt.xlabel('city_mpg')

plt.tight_layout()
plt.savefig('histograms.png')  
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=mpg_df, x='peak_rpm', y='curb_weight', hue='number_of_cylinders')
plt.title('Relationship between Displacement and curb_weight for Origin 2')
plt.xlabel('peak_rpm')
plt.ylabel('curb_weight')
plt.savefig('peak_rpm_curb_weight.plot.png')  # Save the plot as an image file
plt.show()
```

```python
plt.figure(figsize=(10, 8))
vars_of_interest = ['city_mpg', 'number_of_cylinders', 'stroke', 'peak_rpm', 'curb_weight', 'horsepower']
sns.pairplot(data=mpg_df)
plt.suptitle('Relationship of Variables with Kilometer_per_litre')
plt.show()
```

```python
average_weight = mpg_df.groupby('number_of_cylinders')['curb_weight'].mean()

plt.figure(figsize=(8, 6))
sns.barplot(x=average_weight.index, y=average_weight.values)
plt.title('Average Weight based on Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Average Weight')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(data=mpg_df[['peak_rpm', 'wheel_base', 'curb_weight']])
plt.title('Boxplot of peak_rpm, wheel_base, and curb_weight')

plt.subplot(1, 2, 2)
sns.boxplot(data=mpg_df['city_mpg'])
plt.title('Boxplot of city_mpg')

plt.tight_layout()
plt.show()
```
